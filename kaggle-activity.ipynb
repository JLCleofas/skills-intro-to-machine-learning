```python
import sklearn
import sklearn.neighbors
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ahmeduzaki/global-earthquake-tsunami-risk-assessment-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv('earthquake_data_tsunami.csv')
```

```python
df
```

```python
X = df.drop(columns=['Year', 'Month', 'tsunami']).iloc[:626]
y = df['tsunami'].iloc[:626]
```

```python
model = sklearn.neighbors.KNeighborsClassifier()
```

```python
model.fit(X, y)
```

```python
row_626 = df.drop(columns=['Year', 'Month', 'tsunami']).iloc[[626]]
```

```python
prediction = model.predict(row_626)
```
